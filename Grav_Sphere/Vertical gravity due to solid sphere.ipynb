```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
def surface_anomaly_1D(radius, density_diff, depth, x_min, x_max, resolution):
    pi = np.pi
    mass = (4/3)*(pi)*(np.power(radius,3))* density_diff #In Kg
    G = 0.0000000000667259
    x= np.arange(x_min,x_max,resolution).reshape(-1,1)

    temp = (x*x)+(depth*depth)

    r = np.sqrt(temp)

    g_z = (G*mass*depth)/(np.power(r,3))*100000


    plt.figure(figsize=(16,10))
    plt.xlabel("Distance(in m)")
    plt.ylabel("mGal")
    plt.plot(x,g_z,color = 'blue')
    plt.title("Gravity Anomaly due to Sphere : Depth = {} km, Radius = {} m, Density_Diff = {} kg/m3".format(depth,radius,density_diff))
    plt.show()
    return g_z, x 
    
```
