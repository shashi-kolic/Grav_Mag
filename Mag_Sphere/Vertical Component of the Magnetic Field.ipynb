```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
def vertical_mag_field_1D (intensity_of_polarization, depth, radius, density, x_min, x_max, resolution):
    x= np.arange(x_min,x_max,resolution).reshape(-1,1)
    pi = np.pi
    I = intensity_of_polarization
    R = radius
    ro = density
    z = depth
    k = (8*pi*(np.power(R,3))*I)/(3*np.power(z,3))
    temp = (x*x)/(z*z)
    f = (1-0.5*temp)/(np.power((1+temp),5/2))
    plt.figure(figsize=(16,10))
    plt.plot(x,f)
    plt.xlabel("Distance(in m)")
    plt.ylabel("Magnetic Anomaly(oersted)")
    plt.show()
    return f,x
```

```python
vert_mag , x = vertical_mag_field_1D (intensity_of_polarization = 0.5 , depth= 25, radius=25, density=0.25, x_min = -100, x_max=100, resolution=1)
```

```python
vert_mag=pd.DataFrame(vert_mag)
x=pd.DataFrame(x)
magnetic_field = pd.concat([x,vert_mag],axis=1)
magnetic_field.to_csv(r'Vertical_magnetic_field.csv',index = False, header=['X','Vertical_Mag_Field'])
```
